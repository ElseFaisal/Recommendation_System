# Movie Recommendation Systems

```python
import pandas as pd
movieratings = pd.read_csv(r"C:\Users\Faisal\Desktop\Courses\RecommenderSystems\Recommendation_System\Datasets\movieratings.csv",encoding="ISO-8859-1")
```

```python
movieratings.rename(columns={'id':'movieId'},inplace=True)
movieratings.rename(columns={'ï»¿userId':'userId'},inplace=True)
movieratings.head()
```

```python
movieratings1 = movieratings.pivot_table(index=['userId'],columns=['original_title'],values='rating')
```

```python
se7en_ratings = movieratings1['Se7en']
se7en_ratings.head()
```

```python
similarmovies = movieratings1.corrwith(se7en_ratings)
similarmovies = similarmovies.dropna()

df = pd.DataFrame(similarmovies)
df.head(10)
```

```python
similarmovies.sort_values(ascending=False)
```

```python
import numpy as np

moviestats = movieratings.groupby('original_title').agg({'rating':[np.size,np.mean]})
moviestats.head()
```

```python
popularMovies = moviestats['rating']['size'] >= 100
moviestats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = moviestats[popularMovies].join(pd.DataFrame(similarmovies, columns=['similarity']))
df.sort_values(['similarity'],ascending=False).dropna()[:29]
```

# Improve Recommendation System

```python
movieratings2 = movieratings.pivot_table(index=['userId'],columns=['original_title'],values='rating')
```

```python
corrmatrix = movieratings2.corr()
corrmatrix.head()
```

```python
corrmatrix = movieratings2.corr(method='pearson',min_periods=100)
corrmatrix.head()
```

```python
myR = movieratings2.loc[220].dropna()
myR
```

```python
simCand = pd.Series()


for i in range(0,len(myR.index)):
    
    sims = corrmatrix[myR.index[i]].dropna()
    sims = sims.map(lambda x: x * myR[i])
    
    simCand = simCand.append(sims)
    
simCand.sort_values(ascending=False,inplace=True)
simCand.head(10)
    
```

```python
simCand = simCand.groupby(simCand.index).sum()

simCand.sort_values(ascending=False,inplace=True)
simCand.head(10)

    
```

```python
filtered = simCand.drop(myR.index)
filtered.head(10)
```
