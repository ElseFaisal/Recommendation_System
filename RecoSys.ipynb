```python
movieratings.rename(columns={'id':'movieId'},inplace=True)
movieratings.rename(columns={'ï»¿userId':'userId'},inplace=True)
movieratings.head()
```

```python
movieratings1 = movieratings.pivot_table(index=['userId'],columns=['original_title'],values='rating')
```

```python
se7en_ratings = movieratings1['Se7en']
se7en_ratings.head()
```

```python
similarmovies = movieratings1.corrwith(se7en_ratings)
similarmovies = similarmovies.dropna()

df = pd.DataFrame(similarmovies)
df.head(10)
```

```python
similarmovies.sort_values(ascending=False)
```

```python
import numpy as np

moviestats = movieratings.groupby('original_title').agg({'rating':[np.size,np.mean]})
moviestats.head()
```

```python
popularMovies = moviestats['rating']['size'] >= 100
moviestats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = moviestats[popularMovies].join(pd.DataFrame(similarmovies, columns=['similarity']))
df.sort_values(['similarity'],ascending=False).dropna()[:20]
```
